**Q2. Explain the concept of function arguments and parameters in Python?**

**Sol.** A parameter is the variable listed inside the parentheses in the function definition. An argument is the value that is sent to the function when it is called.



````python

# Here a,b are the parameters

def sum(a,b):
  return a+b

print(sum(11,23))
````

**Arguments:**
An argument is a value that is passed to a function when it is called. It might be a variable, value or object passed to a function or method as input. They are written when we are calling the function.

````python
#here 11 and 23 are arguments.

def sum(a,b):
  return a+b

print(sum(11,23))
````

The four types of arguments in Python are:

**Positional arguments:** These are passed to functions in the same order as the parameters are defined.

````python
#Positional Arguments

def add_numbers(a, b):
    return a + b

result = add_numbers(5, 10)
print(result)
````

**Keyword arguments:** These are passed with the parameter name, allowing them to be in any order.

````python
#Keyword arguments

def student(name, age):
    print('Student Details:', name, age)

student('Naman', 21)

student(name='Vaibhav', age=22)

student('Vibhor', age=23)
````

**Default arguments:** These have a default value if no argument is provided.

````python
#Default Arguments

def exponent(base, power=2):
    return base ** power

print(exponent(5))
print(exponent(5, 3))
````

**Variable-length arguments:** These allow passing an arbitrary number of arguments, using args for positional and *kwargs for keyword arguments.

````python
# Variable-Length Arguments

def my_function(*args):
    for arg in args:
        print(arg)

my_function(1, 2, 3)

print("----------------------")

def my_function(**kwargs):
    for key, value in kwargs.items():
        print(key, value)

my_function(name="Vikas", age=20)
````

**Q5. What are iterators in Python and how do they differ from iterables?**

**Sol.**An iterator is an object that represents a stream of data. It implements two methods:

__iter__(): Returns the iterator object itself. This allows the iterator to be used in contexts where an iterable is expected.

__next__(): Returns the next item in the sequence. When there are no more items, it raises the StopIteration exception.

````python
#Example of iterator

list1=["var1","var2","var3","var5","var6"]

fxn= iter(list1)

print(next(fxn))
print(next(fxn))
print(next(fxn))
print(next(fxn))
print(next(fxn))      #last after that will give iterator error.
````

**Iterables:**
An iterable is any object in Python that can return an iterator. It implements the __iter__() method, which returns an iterator object.

*Common Examples:* Lists, tuples, sets, dictionaries, strings, and even files are examples of iterables.

````python
my_list = [1, 2, 3]
for item in my_list:
    print(item)
````

**Key Differences**

1. **Creation:**
   
   **Iterable:** Any object that implements __iter__() method. Examples include lists, tuples, and dictionaries.

  **Iterator:** An object that implements both __iter__() and __next__() methods. Iterators are often created from iterables using the iter() function.

2. **Functionality:**

  **Iterable:** Provides a way to obtain an iterator via __iter__(). It does not directly produce items; instead, it needs to be iterated over.
  
  **Iterator:** Provides a way to access the elements of the sequence one at a time via __next__(). It maintains state and knows the position within the iteration.
3. **Use Case:**

  **Iterable:** Used in for-loops and functions that need an iterable.
  
  **Iterator:** Used when you need to manually control the iteration process and retrieve items one by one.

**Q6.  Explain the concept of generators in Python and how they are defined.**

**Sol.** Generators in Python are a type of iterable that allow you to iterate over a sequence of values without having to store them all in memory at once. They provide a convenient way to handle large data streams or sequences that are computationally expensive to generate. Generators are defined using functions with the yield keyword.

````python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

gen = countdown(5)
for num in gen:
    print(num)
````

**Advantages of generators:**

**1. Memory Efficiency:** Generators generate values on the fly, reducing memory usage compared to storing the entire sequence in memory.

**2. Lazy Evaluation:** Values are computed only when needed, which can lead to performance improvements, especially with large datasets.

**3. Maintain State:** Generators automatically maintain their state between yield statements, making them suitable for iterative processes.

**4. Convenience:** Generators simplify code for iterating over large datasets or streams of data without needing to manage state explicitly.

**Q7. What are the advantages of using generators over regular functions?**

**Sol.** Advantages of Generators Over Regular Loops
There are several advantages to using generators over regular loops. Here are a few of the most important ones:

1.**Memory Efficiency:** One of the biggest advantages of generators is that they are memory efficient. Since a generator yields its values one by one, it does not need to store all of the values in memory at once. This can be especially useful when working with large data sets, as it reduces the risk of running out of memory.

2.**Readability:** Generators can make code easier to read and understand. By breaking up the iteration process into smaller chunks, we can write code that is more concise and easier to follow.

3.**Speed:** Generators can also be faster than regular loops, especially when working with large data sets. This is because they avoid the overhead of creating a list or other data structure to store the results.

````python
# Infinite Counter
def counter():
    count = 0
    while True:
        yield count
        count += 1

# Create an infinite generator
var =counter()

# Print the first 5 values
for _ in range(5):
    print(next(var))
````

**Q9. Explain the purpose and usage of the `map()` function in Python.**

**Sol.**The map() function is a built-in function used for applying a specified function to every item in an iterable (such as a list or tuple) and returning an iterator that produces the results. It’s particularly useful for transforming or processing data in a concise and readable way.

*Syntax*


```
map(function, iterable, ...)

```

**function:** The function that will be applied to each element of the iterable. This function can be a standard function, a lambda function, or any callable.

**iterable:** The iterable (or iterables) whose elements will be processed by the function

````python
#Use of map function

num = [1, 2, 3, 4, 5]

def square(x):
    return x ** 2

print(list(map(square, num)))        # Convert map object to list to view results
````

````python
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

n=5

def fibo(var):
  """function will give fibbonacci series step by step till user given iterations
      var(int)   : integer value given here is 5
      yield(int) :  sum of last 2 elements
  """
  a=0
  b=1
  for i in range(0,var+1):
    yield a
    a,b=b,a+b

fxn_obj=fibo(n)
print(next(fxn_obj))
print(next(fxn_obj))
print(next(fxn_obj))
print(next(fxn_obj))
print(next(fxn_obj))                       #last iterator

````

````python
#Q6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

n=5

def genx(x):
  """ function will return the 2**x one by one using generator
      x(int)   :   values given here is 5
      yield (int) : 2**x
  """
  for i in range(0,x+1):
    yield 2**i

gen_object=genx(n)
print(next(gen_object))              #first iterator    ie 2**0
print(next(gen_object))              #second iterator   ie 2**1
print(next(gen_object))              #third iterator    ie 2**2
print(next(gen_object))              #fourth iterator   ie 2**3
print(next(gen_object))              #fifth iterator    ie 2**4
print(next(gen_object))              #last iterator     ie 2**5


````

````python
#Q7. . Implement a generator function that reads a file line by line and yields each line as a string


textfile = "/content/sample_file"

x = 5

def read_in_lines(file, n):
    with open(file) as fh:
        for i in range(n):
            nlines = fh.readline()
            if nlines:
                yield nlines
            else:
                break

for lines in read_in_lines(textfile, x):
    print(lines)
````
